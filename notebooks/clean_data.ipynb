```python
import pandas as pd
import numpy as np
import os

# Create processed directory if not exists
os.makedirs('../data/processed', exist_ok=True)

# ------------------------------
# 1. Clean Fraud_Data.csv
# ------------------------------
df_fraud = pd.read_csv('../data/raw/Fraud_Data.csv')
print("Fraud data shape:", df_fraud.shape)

# Check missing
print("Missing values:\n", df_fraud.isnull().sum())

# Remove duplicates
initial = len(df_fraud)
df_fraud = df_fraud.drop_duplicates()
print(f"Removed {initial - len(df_fraud)} duplicates")

# Convert date columns
df_fraud['signup_time'] = pd.to_datetime(df_fraud['signup_time'])
df_fraud['purchase_time'] = pd.to_datetime(df_fraud['purchase_time'])

# Save cleaned
df_fraud.to_csv('../data/processed/fraud_data_cleaned.csv', index=False)
print("Saved cleaned fraud data.")

# ------------------------------
# 2. Clean creditcard.csv
# ------------------------------
df_cc = pd.read_csv('../data/raw/creditcard.csv')
print("\nCredit card data shape:", df_cc.shape)
print("Missing values:\n", df_cc.isnull().sum())
print(f"Duplicates: {df_cc.duplicated().sum()}")
# No duplicates expected, but remove if any
df_cc = df_cc.drop_duplicates()
df_cc.to_csv('../data/processed/creditcard_cleaned.csv', index=False)
print("Saved cleaned credit card data.")

# ------------------------------
# 3. IP mapping table (only check structure, no cleaning needed)
# ------------------------------
ip_map = pd.read_csv('../data/raw/IpAddress_to_Country.csv')
print("\nIP mapping shape:", ip_map.shape)
print("Columns:", ip_map.columns.tolist())
ip_map.to_csv('../data/processed/ip_country_cleaned.csv', index=False)
print("Saved IP mapping for later use.")
```
