```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df_cc = pd.read_csv('../data/processed/creditcard_cleaned.csv')
print("Class imbalance:", df_cc['Class'].value_counts())
print("Fraud percentage:", df_cc['Class'].mean() * 100)

# Histograms of Time and Amount
fig, axes = plt.subplots(1,2, figsize=(12,4))
sns.histplot(df_cc['Time'], bins=50, ax=axes[0])
axes[0].set_title('Time distribution')
sns.histplot(df_cc['Amount'], bins=50, ax=axes[1])
axes[1].set_title('Amount distribution')
plt.tight_layout()
plt.show()
```
